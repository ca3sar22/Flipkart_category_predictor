```python
# Now assiging the feature and label as X and y
```

```python
X = df['description']
y = df['product_category_tree']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting X and y into testing and training sets, keeping testing amount of 30 percent and 70 precent of training.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Now here I used Count vectorizer, even tho tfidf vectorizer works better than it as it deals with the 
# overall weightage of a word in a document and helps in dealing with most frequent words. But here my approach was to find
# number of occurance of words and build a vocabulary out of them and for that I thought count vectorizer was enough.
```

```python
count_vec = CountVectorizer()
```

```python
X_train_count = count_vec.fit_transform(X_train)
```

```python
# Here the count vectorizer converted our text data into the form of vectors on the basis of their occurance frequency
```

```python
X_train_count.toarray()
```

```python
X_train_count.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
# Here I have used Tfidf that stands for term frequencyinverse document frequency, tho my first choice was bag of words
# but I avoided it because I thought that we are dealing with the feature which has descriptions in it, it means each description
# can introduce new words and then our vocabulary size would increase and thereby, the length of the vectors would increase too,
# plus it will create more zeros as well which can result in creating a sparse matrix. 
```

```python
# So I went for tfidf here, as it not only measures how frequently a word as occured but also how important that words is in
# the sentences. Therefore, the score of words like 'and', 'or', will get reduced to zero. 
```

```python
tfidf = TfidfTransformer()
```

```python
# Here I converted our count matrix into tfidf representation
```

```python
X_tfidf  = tfidf.fit_transform(X_train_count.astype('U'))
```

```python
X_tfidf.shape
```

```python
from sklearn.svm import SVC
```

```python
#Here I used SVM classifier instead of Multinomial Naive Bayes as Naive Bayes assumes in the posterior probability that 
# every element is Independent of each other. It works well for most of the texts where the length is small. As it is
# a probailistic based classifier so it cannot learn interactions between features because of the independence assumption it
# makes which is rarely true in most real-world applications. And in our casethe description length can vary from lower to 
# higher range as well. So, I went for SVM as it works on the geometric interpretation of the problem. It creates a line or 
# a hyperplane which separates the data into classes. 
```

```python
# According to the SVM algorithm we find the points closest to the hyperplane or line from the classes.These points are called 
# support vectors. Now, we find the distance between them and the support vectors which is called the margin. 
# The hyperplane for which the margin is maximum the SVM tries to make a decision boundary by it.
```

```python
clf = SVC().fit(X_tfidf,y_train)
```

```python
from sklearn.pipeline import Pipeline
```

```python
#Now I used the Pipeline with the purpose of assembling the several steps that I used to validate together.
```

```python
text_clf = Pipeline([('vect',CountVectorizer()),('tfidf',TfidfTransformer()),('clf',SVC()),])
```

```python
# Fitting the model
```

```python
text_clf = text_clf.fit(X_train, y_train)
```

```python
# Predicting
```

```python
pred = text_clf.predict(X_test.values.astype('U'))
```

```python
# Checking the final score with our testing data
```

```python
text_clf.score(X_test.values.astype('U'),y_test)
```
